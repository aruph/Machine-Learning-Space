# 1. WRITE A CODE THAT TAKES A SPEED INPUT FROM A USER IN KM PER HOUR, CONVERTS IT INTO MILES PER HOUR, AND PRINT BOTH SPEEDS TO THE SCREEN

```python
speed_kmph = float(input("Enter speed in kmph: "))
speed_mph = round(0.6214 * speed_kmph, 2)
print ("The equivalent of {} kmph is {} mph".format(speed_kmph, speed_mph))
```

# 7. WRITE A CODE THAT TAKES THE LENGTH AND WIDTH OF A RECTANGE FROM THE USER AND PRINTS THE AREA AND PERIMETER 

```python
length = float(input("Enter any length for rectangle: "))
breadth = float(input("Enter any breadth for rectangle: "))
area = round(length * breadth, 2)
perimeter = round(2*(length + breadth), 2)
print ("The length and breath of the rectangle provided by the user is {} and {}, respectively".format(length, breadth))
print ("\nThe area and perimter based on above dimensions are", area, "and", perimeter, ", respectively")
```
